## Procedures

```python
def send_to_server(url, payload):
    # hit the api with a payload
    return requests.get(url, payload).json()
```

```python
def output_xls(dataset,file,sheet):
    # write out a result to excel
    print 'Starting . . . '
    writer = pd.ExcelWriter(file, engine='xlsxwriter')
    dataset.to_excel(writer, sheet_name=sheet)
    writer.save()
    print 'File output complete.'
    return None
```

```python
def getdata(entity_list):
    # fetch data for multiple entities in sequence
    columns = ['buzz','entity_id','mood','name','negatives','neutrals',
              'positives','type']
    df_results = pd.DataFrame(columns=columns)
    print 'Working ....'
    for item in entity_list:
        payload = {"entities": item}
        # pprint.pprint(send_to_server(summary_url, payload)['results'])
        df_row = pd.DataFrame((send_to_server(summary_url, payload)['results']))
        df_results = df_results.append(df_row)
        print 'Item complete.'
    print 'All Done.'
    return df_results
```

```python
def get_entities(uuid_list):
    # fetch data for multiple uuids in sequence
    columns = ['entity_id','entity_name','relation','uuid','uuid_name']
    df_uuid_results = pd.DataFrame(columns=columns)
    for item in uuid_list:
        send_url = relations_url
        payload = {"entities": item, "names": 'true', 'flatten': 'true', 'inc_header': 'true' }
        # pprint.pprint(send_to_server(send_url,payload)['results'])
        df_row = pd.DataFrame((send_to_server(send_url,payload)['results']))
        df_uuid_results = df_uuid_results.append(df_row)
        print 'Item complete.'
    return df_uuid_results    
```
